#### 2. EDA
- Time 데이터 sec 기준으로 변경
  - 스플릿해서 m*60 + s + p/100
- Rank 데이터 수치형으로 변경
- Kart 데이터 NaN을 신규 카트바디인 '렉키 V1'으로 변경
- Track 데이터 '알수없는 트랙'은 신규 트랙으로 대부분의 유저들의 이해도가 부족할 것으로 판단해서 삭제

```python
import pandas as pd

df = pd.read_csv('preprocessed_data.csv')
df
```

```python
df.info()
```

```python
df.isna().sum().sum()# 맞더라, 결측치랑 리타이어 포함하니까 얼추 맞아
```

```python
for i in range(1,5):
  na_idx = df[f'User_{i}_Time'].isna()
  m, s, p = df.loc[~na_idx, f'User_{i}_Time'].str.split('\'').str
  df.loc[~na_idx, f'User_{i}_Time'] = m.astype(int)*60 + s.astype(int) + p.astype(float)/100
  df.loc[na_idx, f'User_{i}_Time'] = 999.99

df
```

```python
for i in range(1,5):
  na_idx = df[f'User_{i}_Kart'].isna()
  df.loc[na_idx, f'User_{i}_Kart'] = '렉키 V1'
```

```python
df['Track'].unique()
```

```python
df['Track'].value_counts()['알 수 없는 트랙']#매치 수도 적음
```

```python
df = df[df['Track'] != '알 수 없는 트랙']
df = df.reset_index(drop=True)
df
```

```python
kart_data = pd.read_csv('kart_data.csv')
kart_data
```

```python
df_fin = pd.DataFrame()
cnt = 0
for i in range(2784):
  for j in range(1,5):
    df_fin.loc[cnt, 'Match_Id'] = df.loc[i,'Match_Id']
    df_fin.loc[cnt,'Track'] = df.loc[i,'Track']
    cnt += 1
df_fin
```

```python
kart_tmp = kart_data.set_index('Name')
kart_tmp
```

```python
#이름, 랭크, 타임, 카트, 드리프트, 가속력, 코너링, 부스터시간, 부스터충전량
cnt = 0
for i in range(0, 11136, 4):
  for j in range(1,5):
    df_fin.loc[i+j-1, 'Name'] = df.loc[cnt, f'User_{j}_Name']
    df_fin.loc[i+j-1, 'Rank'] = df.loc[cnt, f'User_{j}_Rank']
    df_fin.loc[i+j-1, 'Time'] = df.loc[cnt, f'User_{j}_Time']
    df_fin.loc[i+j-1, 'Kart'] = df.loc[cnt, f'User_{j}_Kart']
    df_fin.loc[i+j-1, '드리프트'] = kart_tmp.loc[df.loc[cnt, f'User_{j}_Kart'], '드리프트']
    df_fin.loc[i+j-1, '가속력'] = kart_tmp.loc[df.loc[cnt, f'User_{j}_Kart'], '가속력']
    df_fin.loc[i+j-1, '코너링'] = kart_tmp.loc[df.loc[cnt, f'User_{j}_Kart'], '코너링']
    df_fin.loc[i+j-1, '부스터시간'] = kart_tmp.loc[df.loc[cnt, f'User_{j}_Kart'], '부스터시간']
    df_fin.loc[i+j-1, '부스터충전량'] = kart_tmp.loc[df.loc[cnt, f'User_{j}_Kart'], '부스터충전량']
  cnt += 1

df_fin
```

#### 특성 설명

- Match Id : 모조리 고유값, 분석할 때는 무조건 드랍.

- Trakc : Time에 직접적인 영향을 끼치는 특성, 다만 Time을 평균값으로 사용한다면 크게 의미없음

- Name : 약간은 타임이랑 관련이 있을 수 있음. 기록이란 건 결국 누군가의 실력이니까, 잘하는 사람이면 기록이 잘 나올 가능성.

- Rank : 판단보류. 기록이 랭크랑 직결되긴 하지만 랭크는 어디까지나 해당 매치 안에서만 작용하는 부분이니까 랭크를 예측하는게 아니라면 그렇게까지 쓸모있을까?

- Time : 타겟이다

- Kart : Time에 직접적인 영향을 주는 특성

- 드리프트, 가속력, 코너링, 부스터시간, 부스터충전량 : Kart의 성능.
