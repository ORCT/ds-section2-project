#### 1. 데이터 전처리
- 전부 한 줄로 되어있는 데이터를 특성별로 분리한다.

```python
import pandas as pd
data = pd.read_table('data_dict_40_100match.txt')
data
```

```python
data.info()
```

```python
index = pd.Series(range(1, 2800))
data.set_index(keys=[index],inplace=True)
data
```

```python
df = pd.DataFrame({'origin_dict':[0]})
df.loc[0,:] = data.columns
data.columns = ['origin_dict']
df = df.append(data,ignore_index=True)
```

```python
df['Match_Id'], df['Match_Data'] = df['origin_dict'].str.split(' : ').str
df
```

```python
df['Match_User_1'], df['Match_User_2'], df['Match_User_3'], df['Match_User_4'] = 0, 0, 0, 0
for i in range(0,2800):
  tmp = df.loc[i,'Match_Data'].split('), ')
  for j in range(len(tmp)):
    df.loc[i, f'Match_User_{j+1}'] = tmp[j]
    
df
```

```python
df.info()
```

```python
df['Track'] = df['Match_User_1'].str.split(',').str[0]
df
```

```python
df['Track'] = df['Track'].str.split('\'').str[1]
df
```

```python
for i in range(1,5):
  df[f'User_{i}_Rank'] = df[f'Match_User_{i}'].str.split(', ').str[1]
  df[f'User_{i}_Kart'] = df[f'Match_User_{i}'].str.split(', ').str[2]
  df[f'User_{i}_Name'] = df[f'Match_User_{i}'].str.split(', ').str[3]
  df[f'User_{i}_Time'] = df[f'Match_User_{i}'].str.split(', ').str[4]
df
```

```python
df.info()
```

```python
for i in range(1,5):
  df[f'User_{i}_Kart'] = df[f'User_{i}_Kart'].str.split('\'').str[1]
  df[f'User_{i}_Name'] = df[f'User_{i}_Name'].str.split('\'').str[1]
  df[f'User_{i}_Time'] = df[f'User_{i}_Time'].str.split('\"').str[1]
df
```

```python
df.info()
```

- 결측치, 중복치, 데이터 타입 변경 등 기본 데이터 정리 및 가공
  - 결측치 : Time 칼럼이 타겟이 될 것이므로 Time 결측치들은 전부 삭제
  - 중복치 : Time이 소수점 둘째자리까지 같고, 트랙도, 이름도, 카트바디까지 같다면 이상치가 확실하므로 지워도 무방함.
  - 데이터 타입 변경 : 수치형이어야 하는 변수들 수치형으로 변경

```python
df_tmp = df.iloc[:,7:]
df_tmp
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(df_tmp.isnull().T,cbar=False)
```

```python
df_tmp = df_tmp.dropna()#결측치 전부 Time이라 드랍
df_tmp.reset_index(inplace=True, drop=True)
df_tmp.isna().sum()
```

```python
df_tmp.duplicated()#기록, 트랙, 카트, 유저 이름 까지 모두 같다면 중복이상치이므로 지워도 무방함
```

```python
df_tmp = df_tmp.drop_duplicates()
df_tmp.info()
```
